# Data Preprocessing

```python
import pandas as pd
```

### Checking for Null columns and duplicates

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(columns=['id'],inplace=True)  # dropping the id column
```

### Converting the dictionary column to a new dataframe

```python
df.loc[0,'amenities']
```

```python
import json
def amens(amenities):

    n = json.loads(amenities)

    m = {key:1 if value else 0 for key,value in n.items()}

    return m
```

```python
df['amenities'] = df['amenities'].apply(lambda x: amens(x))
df.tail(3)
```

```python
new_df = pd.DataFrame(df['amenities'].tolist(), index=df.index)
new_df.drop(columns=['LIFT','GYM','POOL'],inplace=True)
```

```python
new_df.head()
```

```python
df.drop(columns=['amenities'],inplace=True)
```

```python
df.columns
```

```python
df = pd.concat([df, new_df], axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df.describe()
```

```python
df = df[df['property_age']>0]
```

```python
data = df.copy()
```

```python
data['activation_date'] = pd.to_datetime(data['activation_date'])
data.head(2)
```

```python
data['activation_day'] = data['activation_date'].dt.day
data['activation_month'] = data['activation_date'].dt.month
data['activation_year'] = data['activation_date'].dt.year
data.drop(columns='activation_date', inplace=True)
data.head()
```

```python
data['type'].unique()
```

```python
data['type']=data['type'].str.replace('bhk2','BHK2')
data['type']=data['type'].str.replace('bhk3','BHK3')
data['type']=data['type'].str.replace('1BHK1','BHK1')
```

```python
data['type'].unique()
```

```python
data.drop(data.loc[data['cup_board']==48026.0].index, inplace=True)
data.drop(data.loc[data['bathroom']==21.0].index, inplace=True)
data.drop(data.loc[data['balconies']==13.0].index,inplace=True)
data.drop(data.loc[data['property_age']== 400.0].index,inplace=True)
data.drop(data.loc[data['property_age']== 100.0].index,inplace=True)
```

```python
column = ['property_age','bathroom','cup_board','floor','total_floor','balconies','CLUB','CPA','SERVANT','GP','RWH','STP','VP']
for col in column:
    data[col]=data[col].astype(int)
```

```python
data.columns
```

```python
amenities = ['gym','lift','swimming_pool', 'INTERNET','AC', 'CLUB', 'INTERCOM', 'CPA', 'FS', 'SERVANT',
       'SECURITY', 'SC', 'GP', 'PARK', 'RWH', 'STP', 'HK', 'PB', 'VP']
data['Total_amenities'] = data[amenities].sum(axis=1)
```

```python
#data.to_csv('df_for_eda.csv', index=False)
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
columns = ['type','lease_type','furnishing','parking','facing','water_supply','building_type','locality']

for col in columns:
    data[col] = le.fit_transform(data[col])
```

```python
list(data.select_dtypes(include=float).columns)
```

```python
corr = data[['latitude', 'longitude', 'type', 'lease_type', 'property_size', 'property_age', 'furnishing', 'facing', 'floor', 'total_floor',
            'building_type', 'water_supply', 'negotiable', 'cup_board', 'balconies', 'parking', 'bathroom','rent','Total_amenities']].corr()
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 8))  
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

From the above heatmap it can be seen the rent is positively correlated with no.of cupboards,blaconies,bathrooms,floors,property 
size, the type of the property and the total amenities. Thus these are the main features that primarily decide the rent of a property. 
Also the rent is negatively correlated with parking facility and the building type.
